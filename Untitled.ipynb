```python
import json
from pprint import pprint
import matplotlib.pyplot as plt
import numpy as np
# d liu ly t ngy 2018-02-03 n 2018-05-14
with open('update.json') as f:
    data = json.load(f)

trungbinh=0
dem=0
danhsach=[]

for i in range(0,len(data)):
	for j in range(0,data[i]["size"]):
		trungbinh += data[i]["list"][j]["avg_time_on_read"]	
		danhsach.append(data[i]["list"][j]["avg_time_on_read"])


trungbinh /= len(data)*5 #gi tr trung bnh ca d liu

#tnh standard diviation
a = np.array(danhsach)
std_div=np.std(a)


#v ng d liu trung bnh, stmax, stmin
cao=trungbinh+2*std_div
thap=trungbinh-2*std_div
stmax=[]
stmin=[]
listTB=[]
for i in range(0, len(data)*5):
	listTB.append(trungbinh)
	stmax.append(cao)
	stmin.append(thap)
plt.figure(figsize=(19,5))
for i in range(0,len(data)):
	for j in range(0,data[i]["size"]):
		if data[i]["list"][j]["avg_time_on_read"] >cao or data[i]["list"][j]["avg_time_on_read"] <thap:
			plt.plot(dem,data[i]["list"][j]["avg_time_on_read"],"xr-")
		else :
			plt.plot(dem,data[i]["list"][j]["avg_time_on_read"],"bo-")
		
		dem += 1
plt.plot(stmax,color="violet")
plt.plot(stmin,color="violet")
plt.plot(listTB,color="red")
plt.plot(danhsach,color="blue")
plt.show()
```
